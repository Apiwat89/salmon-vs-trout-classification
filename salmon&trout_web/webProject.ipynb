```python
import numpy as np
from flask import Flask, render_template, request
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import load_img, img_to_array
from pyngrok import ngrok

app = Flask(__name__)

# Uncomment this line if run_with_ngrok is needed
# run_with_ngrok(app)

model = load_model('InceptionV3_3.h5')

model.make_predict_function()
ngrok_tunnel = ngrok.connect(5000)

# Print the public URL
print('Public URL:', ngrok_tunnel.public_url)

def predict_label(img_path):
    test_image = load_img(img_path, target_size=(224, 224))
    test_image = img_to_array(test_image) / 255.0
    test_image = test_image.reshape(1, 224, 224, 3)
    result_value = model.predict(test_image)
    print(result_value)
    result = (result_value > 0.5).astype("int32")  # Binary classification
    confidence = round(result_value[0][0] * 100, 2) if result[0][0] == 1 else round((1 - result_value[0][0]) * 100, 2)
    prediction = 'ปลาเทราต์' if result[0][0] == 1 else 'ปลาแซลมอน'
    print(prediction)
    print(result[0][0])
    return confidence, prediction

# routes
@app.route("/", methods=['GET', 'POST'])
def main():
    return render_template("index.html")

@app.route("/about")
def about_page():
    return "Please subscribe to Artificial Intelligence Hub..!!!"

@app.route("/submit", methods=['GET', 'POST'])
def get_output():
    if request.method == 'POST':
        img = request.files['my_image']
        img_path = "static/" + img.filename
        img.save(img_path)
        confidence, prediction = predict_label(img_path)
        return render_template("index.html", prediction=confidence, TEXT=prediction, img_path=img_path)

app.run()
```
